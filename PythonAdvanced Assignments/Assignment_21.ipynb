## Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?

A probability distribution is a mathematical function that describes the probabilities of different outcomes in a random process. While the specific outcomes cannot be predicted with certainty, the probabilities of different outcomes can be determined using a probability distribution.

## Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?

Yes, there is a distinction between true random numbers and pseudo-random numbers.

True random numbers are generated by natural, physical processes that are inherently unpredictable, such as radioactive decay or atmospheric noise. These processes produce numbers that are truly random and have no discernible pattern or correlation.

On the other hand, pseudo-random numbers are generated by algorithms that use mathematical formulas to produce sequences of numbers that appear to be random, but are actually deterministic. Pseudo-random numbers are generated by starting with a seed value and applying a series of mathematical operations to that value to generate the next number in the sequence. While the resulting sequence of numbers appears to be random and unpredictable, it is actually completely determined by the seed value and the algorithm used.

Pseudo-random numbers are considered "good enough" for many applications because they exhibit many of the same statistical properties as true random numbers, such as uniformity and independence. 

## Q3. What are the two main factors that influence the behaviour of a "normal" probability distribution?

The two main factors that influence the behavior of a "normal" probability distribution are the mean and the standard deviation. The mean and standard deviation determine the shape, location, and spread of the normal distribution. The normal distribution is often referred to as the bell curve because of its characteristic bell-shaped curve, which is symmetrical around the mean. 

## Q4. Provide a real-life example of a normal distribution.

A real-life example of normal distribution in real life is the distribution of test scores on a standardized exam, where the majority of students score around the mean score, and fewer students score at the extremes of very low or very high scores.

## Q5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows?

In the short term, a probability distribution may not behave as expected due to randomness, but as the number of trials grows, the behavior of the distribution will tend to become more predictable and consistent, converging to the expected pattern.

The random.shuffle function shuffles the elements of the sequence randomly, meaning that each possible ordering of the elements is equally likely. After the shuffle operation, the original sequence is modified, and its elements are arranged in a random order.

## Q8. What is the relationship between exponentiation and logarithms?

Exponentiation and logarithms are inverse operations of each other. In other words, if we have a number x raised to the power of y, we can find y by taking the logarithm of x to the base of some other number.

More specifically, if we have a number x raised to the power of y, we can write this as:

x^y

Taking the logarithm of both sides to the base of some other number, say b, we get:

log_b(x^y) = y * log_b(x)

In other words, the logarithm of x to the base b raised to the power of y is equal to y times the logarithm of x to the base b.
